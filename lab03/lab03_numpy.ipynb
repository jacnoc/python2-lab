```python
nazwy = np.array(['Laptop', 'Tablet', 'Smartfon', 'Słuchawki', 'Ładowarka','Klawiatura', 'Mysz', 'Monitor', 'Kamera', 'Drukarka'])
ceny = np.array([3499, 1899, 2499, 349, 89, 249, 129, 1499, 599, 899])
ilosci_magazyn = np.array([15, 42, 78, 156, 312, 89, 201, 23, 45, 34])
oceny_klientow = np.array([4.7, 4.2, 4.5, 4.8, 3.9, 4.1, 4.3, 4.6, 3.8, 3.5])
```

```python
print(f"Średnia cena: {ceny.mean()} zł") 
print(f"Najdroższy produkt: {nazwy[ceny.argmax()]} ({ceny.max()} zł)") 
print(f"Najtańszy produkt: {nazwy[ceny.argmin()]} ({ceny.min()} zł)")
```

```python
wartosci_produktow = ceny * ilosci_magazyn 
print(f"Łączna wartość całego magazynu: {wartosci_produktow.sum()} zł")
```

```python
ceny_rabat = ceny * 0.85 
ceny_vat = ceny * 1.23  
powyzej_1000 = (ceny > 1000).sum() 

print(f"Produkty powyżej 1000 zł: {powyzej_1000}") 
```

```python
maska_ocena = oceny_klientow >= 4.5 
print("Produkty z wysoką oceną:", nazwy[maska_ocena]) 

maska_tanie = ceny < 200
print("Produkty poniżej 200 zł:", nazwy[maska_tanie])
```

```python
maska_niski_stan = ilosci_magazyn < 30
print("Niski stan magazynowy:", nazwy[maska_niski_stan]) 

wartosc_niskich = wartosci_produktow[maska_niski_stan].sum()
print(f"Wartość produktów z niskim stanem: {wartosc_niskich} zł")
```

```python
maska_okazja = (oceny_klientow >= 4.5) & (ceny < 1000) 
print("Okazje (wysoka ocena + niska cena):", nazwy[maska_okazja])

srednia_wysoka = ceny[oceny_klientow >= 4.0].mean()
srednia_niska = ceny[oceny_klientow < 4.0].mean()
print(f"Średnia cena (ocena >= 4.0): {srednia_wysoka:.2f} zł")
print(f"Średnia cena (ocena < 4.0): {srednia_niska:.2f} zł")
```
